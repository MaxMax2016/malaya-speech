```python
from glob import glob

files = glob('mic-noise/*.*') + glob('Nonspeech/*.*') + glob('split-intro/*.*')
len(files)
```

```python
!mkdir noise
```

```python
from pydub import AudioSegment
import os
import shutil
from tqdm import tqdm

for filename in tqdm(files):
    folder, file = os.path.split(filename)
    
    new_filename = file.replace(' ', '-').replace('--', '').replace('.mp3', '.wav')
    new_filename = f'noise/{new_filename}'
    
    if filename.endswith('.mp3'):
        audio = AudioSegment.from_mp3(filename)
        audio.set_frame_rate(44100).set_channels(1).export(
            new_filename, format = 'wav'
        )
    else:
        shutil.copy(filename, new_filename)
```
