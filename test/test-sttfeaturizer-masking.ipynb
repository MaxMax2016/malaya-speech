```python
import malaya_speech
import tensorflow as tf
```

```python
featurizer = malaya_speech.featurization.STTFeaturizer()
```

```python
x, sr = malaya_speech.load('../speech/example-speaker/husein-zolkepli.wav')
len(x), sr
```

```python
y = featurizer(x)
y.shape
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (15, 5))
plt.imshow(y.T)
```

```python
y_ = malaya_speech.augmentation.spectrogram.mask_frequency(y, n_freq_mask = 2, 
                                                           width_freq_mask = 16)

plt.figure(figsize = (15, 5))
plt.imshow(y_.T)
```

```python
int(y.shape[0] * 0.1)
```

```python
y_ = malaya_speech.augmentation.spectrogram.mask_time(y, n_time_mask = 1, width_time_mask = int(y.shape[0] * 0.05))

plt.figure(figsize = (15, 5))
plt.imshow(y_.T)
```

```python
y_ = malaya_speech.augmentation.spectrogram.mask_frequency(y)
y_ = malaya_speech.augmentation.spectrogram.mask_time(y_)

plt.figure(figsize = (15, 5))
plt.imshow(y_.T)
```
