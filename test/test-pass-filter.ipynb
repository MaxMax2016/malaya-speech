```python
import malaya_speech
import numpy as np
import random
import malaya_speech.augmentation.waveform as augmentation
```

```python
y, sr = malaya_speech.load('../speech/example-speaker/husein-zolkepli.wav')
y = y[:sr * 10]
len(y), len(y) / sr
```

```python
def downsample(y, sr, down_sr):
    y_ = malaya_speech.resample(y, sr, down_sr)
    return malaya_speech.resample(y_, down_sr, sr)
```

```python
%%time

downsample(y, 16000, random.randint(12000, 16000))
```

```python
%%time

augmentation.random_pitch(y, low = 0.9, high = 1.1)
```

```python
%%time

augmentation.lowpass_filter(
    y, cutoff = random.randint(3000, 6000)
)
```

```python
%%time

augmentation.highpass_filter(
    y, cutoff = random.randint(200, 600)
)
```

```python
%%time

augmentation.bandpass_filter(
    y,
    cutoff_low = random.randint(3000, 6000),
    cutoff_high = random.randint(200, 600),
)
```

```python
%%time

augmentation.random_amplitude_threshold(
    y, low = 1, high = 2, threshold = random.uniform(0.35, 0.8)
)
```

```python
%%time

augmentation.random_pitch(y, low = 0.9, high = 1.1)
```

```python
%%time

augmentation.random_stretch(y, low = 0.7, high = 1.1)
```
