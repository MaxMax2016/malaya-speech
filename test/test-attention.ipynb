```python
import tensorflow as tf
import malaya_speech
from malaya_speech.train import attention
```

```python
layer = attention.MultiHeadAttention(num_heads=2, key_dim=2)
```

```python
# target = tf.keras.Input(shape=[8, 16])
# source = tf.keras.Input(shape=[4, 16])

target = tf.random.normal(shape=[2, 8, 16])
source = tf.random.normal(shape=[2, 4, 16])
```

```python
output_tensor, weights = layer.call(target, source, return_attention_scores=True)
```

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
```
