```python
from glob import glob
from sklearn.model_selection import train_test_split
import json
```

```python
mels = glob('../speech-bahasa/output-yasmin/mels/*.npy')
mels.extend(glob('../speech-bahasa/output-yasmin-parliament/mels/*.npy'))

len(mels)
```

```python
import numpy as np

mel_stat = np.load('../speech-bahasa/yasmin-stats/stats.npy')
mel = np.load(mels[0])
```

```python
%%time

def norm_mean_std(x, mean, std):
    zero_idxs = np.where(x == 0.0)[0]
    x = (x - mean) / std
    x[zero_idxs] = 0.0
    return x

norm_mean_std(mel, mel_stat[0], mel_stat[1])
```
