```python
from collections import defaultdict
import random

vctk_speakers = defaultdict(list)
for f in vctk:
    s = f.split('/')[-1].split('_')[0]
    vctk_speakers[s].append(f)
    
speakers = []
    
for s in vctk_speakers.keys():
    speakers.extend(random.sample(vctk_speakers[s], min(400, len(vctk_speakers[s]))))
    
len(speakers)
```

```python
import malaya_speech

files = dari_pasentran + turki + salina + husein + haqkiem + speakers + random.sample(nepali, 5000) \
+ random.sample(singlish, 5000)
```

```python
import mp
```

```python
import numpy as np
from tqdm import tqdm

def loop(files):
    files, index = files
    for i in tqdm(range(len(files))):
        y, _ = malaya_speech.load(files[i], sr = 22050)
        mel = malaya_speech.featurization.universal_mel(y)
        np.save(f'universal-audio/{index}-{i}.npy', y)
        np.save(f'universal-mel/{index}-{i}.npy', mel)
```

```python
mp.multiprocessing(files, loop, cores = 16, returned = False)
```
