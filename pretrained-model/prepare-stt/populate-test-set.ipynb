```python
import json

with open('bahasa-asr-test.json') as fopen:
    data = json.load(fopen)
```

```python
from tqdm import tqdm
from shutil import copyfile

r = []
for no, f in enumerate(data['X']):
    d = f"test-bahasa/{f.replace('/', '-')}"
    copyfile(f, d)
    r.append({'x': d, 'y': data['Y'][no]})
```

```python
with open('bahasa-asr-testset.json', 'w') as fopen:
    json.dump(r, fopen)
```
