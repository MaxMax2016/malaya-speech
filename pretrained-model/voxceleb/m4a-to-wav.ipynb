```python
import mp
from tqdm import tqdm
from pydub import AudioSegment

def loop(args):
    files = args[0]
    index = args[1]
    for file in tqdm(files):
        file = file[1]
        audio = AudioSegment.from_file(file)
        new_file = file.replace('dev', 'wav').replace('/', '-').replace('-voxceleb-wav-', '/voxceleb/wav/').replace('.m4a', '.wav')
        audio.set_frame_rate(16000).set_channels(1).export(new_file, format="wav")
```

```python
files_index = [(no, f) for no, f in enumerate(files)]
r = mp.multiprocessing(files_index, loop, cores = 120)
```
