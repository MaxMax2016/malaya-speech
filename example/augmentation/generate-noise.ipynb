```python
import malaya_speech
```

```python
import IPython.display as ipd

ipd.Audio('speech/n98.wav')
```

```python
y, sr = malaya_speech.read.flac('speech/7021-79759-0004.flac')
len(y), sr
```

```python
noise, sr = malaya_speech.read.wav('speech/n98.wav')
len(noise), sr
```

```python
noised = malaya_speech.augmentation.waveform.add_noise(malaya_speech.astype.int_to_float(y), 
                                                       malaya_speech.astype.int_to_float(noise),
                                                       factor = 0.2)
```

```python
ipd.Audio(noised, rate = sr)
```
